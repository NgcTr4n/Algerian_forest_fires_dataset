```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import model_selection, metrics, linear_model
import m2cgen as m2c
import numpy as np

# Set display options
pd.set_option('display.max_rows', None)
```

```python
# Read CSV file
df = pd.read_csv('Algerian_forest_fires_dataset.csv', header=1)
df = df.drop([122, 123, 165], axis=0)
df.reset_index(inplace=True)
df = df.drop('index', axis=1)
df
```

```python
# Replace empty values in 'Region'
df.loc[:121, 'Region'] = 'Bejaia'
df.loc[122:244, 'Region'] = 'Sidi-Bel'
# Drop unnecessary columns
df = df.drop(['day', 'month', 'year'], axis=1)
```

```python
# Strip leading/trailing whitespaces in column names and 'Classes'
df.columns = df.columns.str.strip()
df['Classes'] = df['Classes'].str.strip()

# Display information about the DataFrame
df.info()
```

```python
# Convert specified columns to numeric
Column_list = ["Temperature", 'RH', 'Ws', 'Rain', "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]
for i in Column_list:
    df[i] = pd.to_numeric(df[i], errors='coerce')  # Coerce invalid values to NaN

# Drop rows with NaN values
df = df.dropna()
```

```python
# Extract numeric and categorical columns
Numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
categorical_col = set(df.columns).difference(set(Numeric_columns))

# Visualization of the DataFrame
fig, ax = plt.subplots(2, 5, figsize=(15, 3))
for i in range(2):
    for j in range(5):
        c = (i * 5) + j
        column = df.columns[c]
        ax[i, j].scatter(df[column], df['Classes'])
        ax[i, j].set_xlabel(column)
        ax[i, j].set_ylabel('Classes')
plt.tight_layout()
plt.show()
```

```python
# Mapping classes to numeric values
class_mapping = {'not fire': 0, 'fire': 1}
df['classes'] = df['Classes'].map(class_mapping)

# Verify the changes
print(df[['Classes', 'classes']])
df.head()
```

```python
 #Create dataset from DataFrame
X = df.loc[:, 'Temperature':'FWI']
y = df['classes']

# Split the dataset
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.2, random_state=1234)
X_train, X_val, y_train, y_val = model_selection.train_test_split(X_train, y_train, test_size=0.25, random_state=1234)
print('\n(X_train, y_train)', X_train, '\n', y_train)
print('\n(X_val, y_val) : ', X_val, '\n', y_val)
print('\n(X_test, y_test) : ', X_test , '\n', y_test)
```

```python
# Linear Regression
# Training
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)

print("------- LinearRegression -------")

# validation
from sklearn import metrics
y_val_hat = model.predict(X_val)
Linear_R2_Validation = metrics.r2_score(y_val,y_val_hat)
print(Linear_R2_Validation)
#evaluation
y_test_hat = model.predict(X_test)
Linear_R2_Evaluation = metrics.r2_score(y_test,y_test_hat)
print(Linear_R2_Evaluation)

# visualization of prediction error
import matplotlib.pyplot as plt
fig, ax = plt.subplots (ncols=2, figsize =(8, 4))
metrics. PredictionErrorDisplay.from_predictions (y_val, y_val_hat, kind='actual_vs_predicted', ax=ax[0])
ax[0].set_title ('Validation set')
metrics.PredictionErrorDisplay.from_predictions (y_test, y_test_hat, kind='actual_vs_predicted', ax=ax[1])
ax[1].set_title ('Testing set')
plt.tight_layout()
plt.show ()
```
